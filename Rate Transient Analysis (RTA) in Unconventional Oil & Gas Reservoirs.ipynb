# Feature Engineering

* We will normalize the data frame

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
scaler.fit(df)
df_scaled = scaler.transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=[df.columns[0], df.columns[1], df.columns[2], df.columns[3], df.columns[4]])
```

```python
df_scaled
```

```python
y_scaled = df_scaled['Aroot(K)']
x_scaled = df_scaled.drop(['Aroot(K)'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
seed = 1000
np.random.seed(seed)
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size = 0.3)
X_train
```

```python
from sklearn.linear_model import LinearRegression
np.random.seed(seed)
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_, x_scaled.columns, columns = ['Coefficient'])
coeff_df
```

```python
f, ax = plt.subplots(figsize = (10, 8))
y_pred = lm.predict(X_test)
ax.scatter(y_test, y_pred, c='b')
ax.set(xlabel = 'Testing actuals', ylabel = 'Testing predictions', title = 'Actual vs Predicted, Testing Data Set (30% of the data)')
plt.show()
```

```python
from sklearn import *
test_set_r2 = metrics.r2_score(y_test, y_pred)
print('Testing r^2:', round(test_set_r2,2))
```

```python
print('MAE: ', round(metrics.mean_absolute_error(y_test, y_pred),5))
print('MSE: ', round(metrics.mean_squared_error(y_test, y_pred),5))
print('RMSE: ', round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),5))
```

```python
sns.distplot((y_test-y_pred))
```
